```python
def sample_alpha(n, k, alpha):
    M = matrix ([[0 if random () <= 1/(1+alpha) else 1 for j in range(1, k+1)] for i in range(1, n+1)])
    P, Q = RSK(M, insertion = RSK.rules.dualRSK)
    return list(P.shape())
```

```python
n=100
k=300
alpha=1/2.0
rows=sample_alpha(n,k,alpha)

# create a figure and axes
fig = plt.figure(figsize=(6,6))
ax1 = plt.subplot(1,1,1)   
ax1.set_xlim(( 0, k))            
ax1.set_ylim((0, n))

ax1.set_aspect('equal','box')
plt.grid()
diagdata=([x for x in rows for _ in range(2)]+[0],
[0]+[y for y in range(1,len(rows)+1) for _ in range(2)])
plt.plot(*diagdata)
plt.fill_between(*diagdata,alpha=0.4)
plt.show()       
```
