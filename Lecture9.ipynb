```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
n=128
k=16384
rows=sample_diagram_schur_weyl(n,k)

# create a figure and axes
fig = plt.figure(figsize=(6,6))
ax1 = plt.subplot(1,1,1)   
ax1.set_xlim(( 0, np.sqrt(k)*3))            
ax1.set_ylim((0, n))

ax1.set_aspect('equal','box')
plt.grid()
diagdata=([x for x in rows for _ in range(2)]+[0],
[0]+[y for y in range(1,len(rows)+1) for _ in range(2)])
plt.plot(*diagdata)
plt.fill_between(*diagdata,alpha=0.4)
plt.show()       
```

```python
# create a figure and axes
fig = plt.figure(figsize=(12,12))
ax1 = plt.subplot(1,1,1)   
ax1.set_aspect('equal','box')
ax1.set_ylim([0,10])
ax1.set_xlim([-1,9])
c=np.sqrt(k)/n
plt.plot([-1,0,9,8,-1],[1,0,9,10,1])

#n=np.sqrt(16384)
plt.plot([(x-i-dx)/n for (i,x) in enumerate(rows) for dx in (0,1)],
         [(x+i+dx)/n for (i,x) in enumerate(rows) for dx in (0,1)])
xs=np.arange(-1,c*(c+2),1/1000)
rho=np.nan_to_num(1/np.pi*np.arccos(1/2*(xs/c+0j+c)/np.sqrt(xs+1)))
ys=1+scipy.integrate.cumtrapz(1-2*rho, xs,initial=0)
#ys=2/np.pi*(xs*np.arcsin(xs/2)+np.sqrt(4-xs*xs))
with matplotlib.rc_context({'lines.linewidth': 0.5, 'lines.linestyle': '-'}):
    plt.plot(xs,ys,color='k')
plt.show()
```

Let's scale the diagram in such a way, that number of boxes is fixed. It means that we divide both x and y by c. 
We then plot for c=0.05, 0.5, 1, 2.

```python
# create a figure and axes
fig = plt.figure(figsize=(12,12))
ax1 = plt.subplot(1,1,1)   
ax1.set_aspect('equal','box')
ax1.set_ylim([0,6])
ax1.set_xlim([-2,4])
c=np.sqrt(k)/n
plt.plot([-2,0,4],[2,0,4])

for c,s in [(0.05,'solid'),(1/2,'dotted'),(1,'dashdot'),(2,'dashed')]:
#n=np.sqrt(16384)
#plt.plot([(x-i-dx)/n for (i,x) in enumerate(rows) for dx in (0,1)],
#         [(x+i+dx)/n for (i,x) in enumerate(rows) for dx in (0,1)])
    xs=np.arange(-1,c*(c+2),1/1000)
    rho=np.nan_to_num(1/np.pi*np.arccos(1/2*(xs/c+0j+c)/np.sqrt(xs+1)))
    ys=1+scipy.integrate.cumtrapz(1-2*rho, xs,initial=0)
#ys=2/np.pi*(xs*np.arcsin(xs/2)+np.sqrt(4-xs*xs))
    with matplotlib.rc_context({'lines.linewidth': 1, 'lines.linestyle': s}):
        plt.plot(xs/c,ys/c,color='b')
plt.show()
```
